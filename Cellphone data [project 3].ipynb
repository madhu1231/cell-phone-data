# Linear Regression

```python
y=phone.Price
```

```python
y
```

```python
x=phone.iloc[:,3:14]
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model_lnr=LinearRegression()
```

```python
model_lnr.fit(x_train,y_train)
```

```python
predict_lnr=model_lnr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(predict_lnr,y_test)
```

# KNN Regressor

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
model_knr=KNeighborsRegressor()
```

```python
model_knr.fit(x_train,y_train)
```

```python
predict_knr=model_knr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(predict_knr,y_test)
```

# Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model_dtr= DecisionTreeRegressor()
```

```python
model_dtr.fit(x_train,y_train)
```

```python
predict_dtr=model_dtr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(predict_dtr,y_test)
```

# Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model_rfr=RandomForestRegressor()
```

```python
model_rfr.fit(x_train,y_train)
```

```python
predict_rfr=model_rfr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(predict_rfr,y_test)
```
